```python
#install the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("phishing.csv")
data
```

```python
data.shape
```

```python
data.info()  #gives the info about the data
```

```python
#Gives the statistical analysis
data.describe(include="all")  # we dont have any object column here
```

```python
#Check how many missing values are in each column
data.isna().sum()
```

```python
#To check all unique values in each column
for col in data.columns:
    print(f"{col}:{data[col].unique()}")
```

```python
plt.figure(figsize=(15,50), facecolor='white')
plotnumber =1

for column in data.drop(['Result'],axis=1):
    ax = plt.subplot(10,3,plotnumber)
    sns.countplot(x=data[column].astype(str),palette="Set1")
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(15,50), facecolor='white')
plotnumber =1

for column in data.drop(['Result'],axis=1):
    ax = plt.subplot(10,3,plotnumber)
    data[column].value_counts().plot.pie(autopct="%1.1f%%")
    plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```
